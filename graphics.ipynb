```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_excel('leitura.xlsx', sheet_name='dados')
df.head()
```

```python
x = df.tempo
y = df.temperatura
plt.plot(x,y)
```

```python
medias_temperatura = df.groupby('periodo').temperatura.mean()
medias_temperatura
```

```python
x = medias_temperatura.index
y = medias_temperatura.values
plt.bar(x,y)
```
